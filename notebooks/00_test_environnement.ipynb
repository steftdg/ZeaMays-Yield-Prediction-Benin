```python
# Cellule 3 — test rapide pour confirmer que tout fonctionne
import matplotlib
data_test = pd.DataFrame({
    "departement": ["Zou", "Collines", "Borgou", "Atacora"],
    "rendement_t_ha": [2.1, 1.9, 1.3, 1.1],
    "precipitation_mm": [1100, 980, 870, 750]
})

plt.figure(figsize=(6, 4))
plt.bar(data_test["departement"], data_test["rendement_t_ha"], color="steelblue")
plt.title("Test — rendement fictif par département")
plt.ylabel("Rendement (t/ha)")
plt.tight_layout()
plt.savefig("../reports/figures/test_environnement.png", dpi=150)
plt.show()
print("Graphique généré et sauvegardé.")
```

```python
# Cellule 4 — test XGBoost minimal
X = data_test[["precipitation_mm"]]
y = data_test["rendement_t_ha"]

model = xgb.XGBRegressor(n_estimators=10, random_state=42)
model.fit(X, y)
pred = model.predict(X)
print("XGBoost fonctionne correctement.")
print(f"Prédictions test : {pred.round(2)}")
```
